````python
dates_list=list(categoricals["Effective To Date"].iloc[0:])
dates_list
````

````python
year=str(dates_list[1])[:1]
year
````

````python
month=[]
for i in dates_list:
    year_var=str(i)[:1]
    month.append(year_var)
print(set(month))
````

````python
categoricals["Effective To Date"]=month
````

````python
categoricals.head()
````

````python
month_categories = {"1" : "January", "2" : "February"}
def clean_dates(x):
    if x[0] in list(month_categories.keys()):
        return month_categories[x[0]]
    else:
        return np.NaN
categoricals['Effective To Date'] = list(map(clean_dates, categoricals["Effective To Date"]))
````
